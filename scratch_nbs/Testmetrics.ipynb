```python
import json

import pandas as pd
from pathlib import Path

pd.set_option("max_colwidth", 300)
from pprint import pprint
import re
```

```python
java_files = sorted(Path("test_data/java/").glob("**/*.gz"))
```

```python
columns_long_list = [
    "repo",
    "path",
    "url",
    "code",
    "code_tokens",
    "docstring",
    "docstring_tokens",
    "language",
    "partition",
]

columns_short_list = ["code_tokens", "docstring_tokens", "language", "partition"]


def jsonl_list_to_dataframe(file_list, columns=columns_long_list):
    """Load a list of jsonl.gz files into a pandas DataFrame."""
    return pd.concat(
        [
            pd.read_json(f, orient="records", compression="gzip", lines=True)[columns]
            for f in file_list
        ],
        sort=False,
    )
```

```python
java_df = jsonl_list_to_dataframe(java_files)
```
